```python
def init_driver(gecko_driver='', user_agent='', load_images=True, is_headless=False):
    firefox_profile = webdriver.FirefoxProfile()
    
    # set_preference in the geckdriver in the documentation
    firefox_profile.set_preference('dom.ipc.plugins.enabled.libflashplayer.so', False) # disable flash plyer
    
    firefox_profile.set_preference("media.volume_scale", "0.0") # mute any volume
    firefox_profile.set_preference("dom.webnotifications.enabled", False) # disable notification

    if user_agent != '':
        firefox_profile.set_preference("general.useragent.override", user_agent)

    # load image    
    if not load_images:
        firefox_profile.set_preference('permissions.default.image', 2)

    options = Options()
    options.headless = is_headless
    
    driver = webdriver.Firefox(options=options,
                               executable_path=f'{current_path}/{gecko_driver}',
                               firefox_profile=firefox_profile)
    
    return driver

```

# [Popular Quotes](https://www.goodreads.com/quotes?page=1) with all labels

```python
MAX_PAGE_NUM = 101
all_qutes = []
all_labels = []

for i in range(1,MAX_PAGE_NUM):
    url = f'https://www.goodreads.com/quotes?page={str(i)}'
    driver.get(url)
        
    quotes_section = driver.find_elements_by_css_selector('.leftContainer .quote')
    for each_quote in quotes_section:

        try:
            # check if the label exist, if it we will add the quote and label
            if each_quote.find_element_by_css_selector('.quote .greyText').text:
                # adding each quote in the page
                quote_words_with_author = each_quote.find_element_by_css_selector('.quoteText').text

                #take just the quote
                quote = quote_words_with_author.split('“')
                quote = quote[1].split('”')[0]
                all_qutes.append(quote)
                print(quote)

                # adding all labels of each quote in the page
                labels_of_quote = each_quote.find_element_by_css_selector('.quote .greyText').text

                # take just labels
                labels = re.sub('tags: ', '', labels_of_quote).split(',')
                print(labels)
                all_labels.append(labels)
                print('-'*50)
            
        except:
            pass
```

```python
len(all_qutes), len(all_labels)
```

```python
dict = {'quotes':all_qutes,
        'tags': all_labels
       } 
  
popular_quotes_df = pd.DataFrame(dict) 
popular_quotes_df.to_csv('popular_quotes.csv', index = False)
popular_quotes_df.head()
```
